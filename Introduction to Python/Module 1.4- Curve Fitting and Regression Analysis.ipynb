#Curve Fitting and Regression Anaylsis

Very similar to other data processing programs, Python is able to interpolate information from the plotted data. To do this we will use the `stats` tool from the `scipy` library. Within the `stats` tool, we can call the `linregress` function to prefrom our linear regression on our data. The `linregress` function will allow us to find the best fit for our curve and in doing so allow us to able to take the slope of our model. If you wish to take a look all the `stats` function has to offer please follow the provided [hyperlink.](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html?highlight=slope)

```python
 from scipy import stats
```

Let us create some random data to play with.

```python
rng = np.random.default_rng()
#line above imports random data
```

To get a hang of our new tool set, lets start with somewhat linear data. Take a look at the printed data to get a feel for 

```python
x = rng.random(10)
y = 1.6*x + rng.random(10) # y = mx + b 
print("x Data Points:", x) #Print the data
print("y Data Points:", y)
```

```python
#Define and preform the linregress function
reg = stats.linregress(x, y)

#Plot the data with our fitted line
plt.plot(x, y, 'o', label='original data')
plt.plot(x, reg.intercept + reg.slope*x, color = 'r', label='fitted line')
plt.legend()
plt.show() 

#Print out the slope, R-squared and y-intercept value for our best fit
print("Slope:", reg.slope)
print("R-squared:", reg.rvalue**2)
print("y-intercept:", reg.intercept)
```
