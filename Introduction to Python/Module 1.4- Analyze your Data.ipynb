# Reference Guide

## Getting Python to talk to you

The `print()` command is our primary way of asking Python for information. You can place any text (in 'quotation marks') or variable names that you want inside the parentheses and Python will display this information on the scren when you run the code cell.

## Your database

Your `database` variable contains all the information from your spreadsheet. You can retrieve information using the names you entered at the top of each column. For example, if you have a column called `x` and another called `y`, you would access these columns using `database['x']` and `database['y']`. Note the quotation marks around the column names. Each of these sets of data is called an **array**.

If you want to access specific elements from an array, you can use `[square brackets]` with a number. `database['x'][0]` accesses the 0th element in the array, `database['x'][1]` accesses the 1st, etc. If you want a range of values, you can use `:` like so: `database['x'][1:14]` gives you elements `1` through `13`.

## Graphing your data

You can graph your data in two ways:
1. If you need a quick graph and only care about one set of data, you can use `database.plot('x','y')`, with column names for `x` and `y`.
2. If you want more than one set of data in the same figure, you can use `plt.plot(database['x'],database['y']`.

Once your graph is created, you can adjust options like `plt.xlabel('x-axis label')`, `xlim(xmin,xmax)`, and their corresponding commands for the y-axis.

## Getting stats from your data

For a column in `database`, you can calculate all sorts of statistics from the `numpy` library:

1. `np.average( database['x'] )` gives you the average of all your `x` values.
2. `np.std( database['x'] )` gives you the standard deviation of your `x` values.
3. `np.min( database['x'] )` gives you the minimum of your `x` values.
4. `np.max( database['x'] )` gives you the maximum of your `x` values.

## Making calculations from your data

You can use your columns in `database` to carry out all sorts of calculations.

* The operators `+, -, *, /` work just like you'd expect.
* `x**y` will raise `x` to the `y` power.
* `numpy` has all sorts of functions you might be interested in: `np.sin(), np.cos(), np.exp(), np.arcsin()...` You can usually guess the names, but a Google search can show you the function you want if your guess doesn't work.

## Defining a function

We can **define a function** using Python's `def` structure:
```
def FunctionName(inputs):
  do so math here
  return outputs
  ```

* `FunctionName` is the function's name.
* `inputs` is any variables you want the function to know, separated by commas. 
* The section `do some math here` carries out what you want the function to do.
* `return` is followed by all the outputs you want from this function.
* The indentation tells Python which lines are inside the function. Once you unindent, Python assumes you're no longer writing `FunctionName`.
* You **call the function** elsewhere using `FunctionName` and giving the numerical values for each input. For example, if you defined a function `z(x,y)`, you could call it using `z(x=1,y=3)`.
