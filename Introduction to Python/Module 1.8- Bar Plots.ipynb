# Bar Plots

The most popular way to graph data in Python is with the Matplotlib library. 

Run the code cell below and describe the output it produces.

```python
%precision 3 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 

Names = np.array(['Megan','Julie','Sam'])
Heights = np.array([5.5,5.75,5.4])

# Create a figure.
MyFigure, MyPlot = plt.subplots()

# Plot data.
MyPlot.bar(Names,Heights)
```
