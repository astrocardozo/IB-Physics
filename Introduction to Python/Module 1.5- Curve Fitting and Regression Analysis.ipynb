#Curve Fitting and Regression Anaylsis

Very similar to other data processing programs, Python is able to interpolate information from the plotted data. To do this we will use the `stats` tool from the `scipy` library. Within the `stats` tool, we can call the `linregress` function to prefrom our linear regression on our data. The `linregress` function will allow us to find the best fit for our curve and in doing so allow us to able to take the slope of our model.

```python
import matplotlib.pyplot as plt
from scipy import stats
#Notice we have already imported matplotlib. This was not necessary
```

Let us create some random data to play with.

```python
rng = np.random.default_rng()
#line above imports random data
```

To get a hang of our new tool set, lets start with somewhat linear data. Take a look at the printed data to get a feel for 

```python
x = rng.random(10)
y = 1.6*x + rng.random(10)
print(f"x Data Points: {x}")
print(f"y Data Points: {y}")
```

```python
#Define and preform the linregress function
res = stats.linregress(x, y)

#Plot the data with our fitted line
plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.legend()
plt.show() 

#Print out the slope and R-squared value for our best fit
print(f"Slope: {res.slope:.3f}")
print(f"R-squared: {res.rvalue**2:.3f}")
```
