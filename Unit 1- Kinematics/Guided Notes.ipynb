# Studying motion in a straight line is the first step to understanding physics. 

* We live in a universe of motion. From the planets to atoms, everything has motion.  

 

* Linear Motion (also called one-dimensional motion) is motion in a straight line. 

 

* For many scenarios we can model systems as objects: For rigid systems we can use any point to represent motion, we use the center of mass. 

 

* All motion is relative to some frame of reference 

 

* Position, Displacement, Average Speed and Average Velocity  

 

* Speed is the distance covered per unit time. 

    * Scalar quantity 

    * Instantaneous speed 

 

* Distance is the total length of the path traveled between two positions 

    * Scalar quantity  

 

* Time is change, or the interval over which change occurs. 

 

* Constant velocity means moving at a constant speed without changing direction. 

    * An object has constant velocity if $v_{avg}$ has the same value for any time interval. If the velocity is constant, the object always moves in the same direction and maintains a steady speed. 

# Velocity is the rate of change of position, and acceleration is the rate of change of velocity 

* An observer in a reference frame can describe the motion of an object using such quantities as position, displacement, distance, velocity, speed, and acceleration. 

* Displacement, velocity, and acceleration are all vector quantities.  
* Displacement is change in position.  

    * Vector 
    * Not to be mistaken with distance. 

    * Must have defined coordinate system  

    * A choice of reference frame determines the direction and the magnitude of quantity. 

* Velocity is the change in position per unit time. 

    * Vector 

    * A choice of reference frame determines the direction and the magnitude of quantity. 

    * The fundamental forces determine both the structure of objects and the motion of objects.  

* Average Acceleration is the rate of change of velocity 

    * Vector 

    * Proportional to the net force on an object 

*  All vectors can be either positive or negative depending on reference frame 

 

* We'll build on the ideas of linear motion in later units to understand more complex kinds of motion, such as motion in 2 or 3 dimensions  
 

# Motion Diagrams and Graphs

* How to Draw a Motion Diagram 

    * Set coordinate system 

    * Depict motion is set segments  

    * Use dots to simplify the diagrams  

    * They can be one, two or even three dimensional 
* Common graphs in physics  

    * Constants 

    * Linear  

    * Inverse  

    * Parabolic 
Uses for graphs 

    * Extrapolate  

        * Predict values outside the range of the graph 

    * Interpolate 

        * Predict values inside the range of the graph  

    * Visualize relationships between variables  

* Variables, Slope and Area under the curve 

    * When two physical quantities are plotted against one another in such a graph, the horizontal  axis (x-axis) is usually considered to be an independent variable and the vertical axis (y-axis) a dependent variable. 

    * Slope of the graph yields relationship between independent and dependent variable. 

    * Slope goes down the Totem Pole 

    * Area goes up the Totem Pole 
* Motion diagrams and x-t graphs are important tools for interpreting what happens during linear motion.
