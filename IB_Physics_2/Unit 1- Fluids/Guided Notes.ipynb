```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def Pressure(F,A):
    P = F/A
    return P

F = 1.22E5
A = np.linspace(0,100,25)
y = Pressure(F,A)

plt.scatter(A,y)
plt.xlabel('Cross Sectional Area (m^2)')
plt.ylabel('Pressure (Pa)')
```

```python
def Pressure(F,A):
    P = F/A
    return P

F = 1.22E5
A = np.linspace(0,100,10)
y = Pressure(F,A)

plt.scatter(1/A,y)
plt.xlabel('1/Cross Sectional Area (m^-2)')
plt.ylabel('Pressure (Pa)')
```

```python
def Pressure(rho,g,h):
    P = rho*g*h
    return P


h = np.linspace(0,100,10)
rho = 1E3
g = 9.81
y = Pressure(rho,g,h)

plt.scatter(h,y)
plt.xlabel('Height (m)')
plt.ylabel('Pressure (Pa)')
```

```python
def Pressure(P1,rho,g,h):
    P = P1 + rho*g*h
    return P

P1 = 1.01E5
h = np.linspace(0,100,10)
rho = 1E3
g = 9.81
y = Pressure(P1,rho,g,h)

plt.scatter(h,y)
plt.xlabel('Height (m)')
plt.ylabel('Pressure (Pa)')
```

```python
def Pressure(P1,rho,g,h):
    P = P1 + rho*g*h
    return P

h = np.linspace(0,100,100)
x = np.linspace(0,100,100)
x,h = np.meshgrid(x,h)
P1 = 1.01E5
rho = 1E3
g = 9.81
z = Pressure(P1,rho,g,h)
ax = sb.heatmap(z)
plt.xlabel('Range (m)')
plt.ylabel('Depth (m)')
```
